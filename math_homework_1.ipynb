```python
a = np.array([20,20,20])
b = np.array([0,0,-20])
```

```python
def get_lenght(array):
    res = 0
    for i in range(array.shape[0]):
        res += array[i]**2
    return math.sqrt(res)
```

```python
len_a = get_lenght(a)
len_a
```

```python
len_b = get_lenght(b)
len_b
```

### Окружность

### Эллипс

```python
a = 10
b = 6
x = np.linspace(-a, a, 1000)
y1 = ((1 - x**2/a**2)*b**2)**0.5
y2 = -((1 - x**2/a**2)*b**2)**0.5
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-max(a,b)-1,max(a,b)+1)
plt.ylim(-max(a,b)-1,max(a,b)+1)
plt.grid(True)
```

### Гипербола

```python
a = 9
b = 7
y = np.linspace(-2*max(a,b), 2*max(a,b), 1000)
x1 = ((y**2/b**2 + 1)*a**2)**0.5
x2 = -((y**2/b**2 + 1)*a**2)**0.5
plt.plot(x1, y)
plt.plot(x2, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-2*max(a,b), 2*max(a,b))
plt.ylim(-2*max(a,b), 2*max(a,b))
plt.grid(True)
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
# Реализуется отрисовка трехмерного пространства
fig = figure()
ax = Axes3D(fig)
# X и Y - массивы для создания сетки
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
# Подготавливается сетка
X, Y = np.meshgrid(X, Y)
# Плоскость
Z = 1*X + 1*Y
Z1 = 1*X + 1*Y + 5
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)
show()
```
